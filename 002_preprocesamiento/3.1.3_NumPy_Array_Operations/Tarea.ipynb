```python
import numpy as np
```

```python
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
array3 = np.array([7, 8, 9])

print(array1)
print("\n")
#Muestra las filas y columnas del array
print(array1.shape)
#Tama√±o del array
print(array1.size)
#dimensiones array
print(array1.ndim)
```

```python
#Suma cada celda
arraySuma = array1 + array2 + array3
print(arraySuma)

#Resta cada celda
arrayResta = array1 - array2 - array3
print(arrayResta)

#Multiplica cada
arrayMult = array1 * array2 * array3
print(arrayMult)

# el 1er array es la base, 2do array exponente
arrayExpone = array1 ** array2
print(arrayExpone)

#divide cada celda del array
arrayDiv = arrayMult / arraySuma
print(np.round(arrayDiv, 2))
```
