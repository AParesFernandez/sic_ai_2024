# Series

- Una serie es una forma de arrays unidimensionales enumeradas secuencialmente

a) Como es un array, todos los elementos de la serie deben pertenecer a un tipo de datos.
b) Se pueden crear series simples a partir de secuencias como listas, tuplas y matrices.
c) d) La palabra "serie" es singular Y plural. Su origen latino, 'serere', significa unir o conectar.
e) El valor predeterminado de la serie es un índice entero. La etiqueta del primer elemento es 0, la etiqueta del segundo elemento es 1 y aumenta de esta manera.
f) El valor del atributo de la serie es la lista de todos los valores de la serie.
g) El valor del atributo del índice significa un índice de una serie.
h) El valor del atributo de index.values ​​es un array de todos los valores del índice.

# DataFrame

- Un DataFrame es una estructura de datos bidimensional que almacena varios tipos de datos (letras, números enteros, valores de punto flotante, datos categóricos, etc.) en una columna.
- Cada tabla tiene tres columnas con una etiqueta de columna. Las etiquetas de las columnas son "Nombre", "Edad" y "Sexo".
- La columna "Nombre" consta de datos de texto en los que cada valor es una cadena, la columna "Edad" es un número plateado y la columna "Sexo" son datos de texto.
- Es similar a la representación de una tabla de datos en una hoja de cálculo.

### Series Operations

```python
student1 = pd.Series({'kor':100, 'eng':80, 'math':90})
student1
```

```python
percentage = student1/200
percentage
```

```python
student1 = pd.Series({'kor':100, 'eng':80, 'math':90})
student2 = pd.Series({'math':80, 'kor':90, 'eng':80})

addition = student1 + student2
subtraction = student1 - student2
multiplication = student1 * student2
division = student1 / student2
```

```python
result = pd.DataFrame([addition, subtraction, multiplication, division],
                      index=['addition', 'subtraction', 'multiplication', 'division'])
result
```

```python
student1 = pd.Series({'kor':np.nan, 'eng':80, 'math':90})
student2 = pd.Series({'math':80, 'kor':90})

addition = student1 + student2
subtraction = student1 - student2
multiplication = student1 * student2
division = student1 / student2
```

```python
result = pd.DataFrame([addition, subtraction, multiplication, division],
                      index=['addition', 'subtraction', 'multiplication', 'division'])
result
```

```python
student1 = pd.Series({'kor':np.nan, 'eng':80, 'math':90})
student2 = pd.Series({'math':80, 'kor':90})

sr_add = student1.add(student2, fill_value=0)
sr_sub = student1.sub(student2, fill_value=0)
sr_mul = student1.mul(student2, fill_value=0)
sr_div = student1.div(student2, fill_value=0)
```

```python
result = pd.DataFrame([sr_add, sr_sub, sr_mul, sr_div],
                      index=['addition', 'subtraction', 'multiplication', 'division'])
result
```
