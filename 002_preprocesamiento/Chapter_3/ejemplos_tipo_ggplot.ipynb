## Configuración básica con estilo ggplot en matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np

# Usar el estilo ggplot en matplotlib
plt.style.use('ggplot')

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el gráfico
plt.plot(x, y)
plt.title('Ejemplo de Gráfico con Estilo ggplot')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
````

## Gráfico de dispersión con seaborn y estilo ggplot

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Datos de ejemplo
np.random.seed(0)
x = np.random.rand(100)
y = np.random.rand(100)
data = pd.DataFrame({'x': x, 'y': y})

# Configurar seaborn para usar el estilo ggplot
sns.set(style="whitegrid")

# Crear el gráfico de dispersión
sns.scatterplot(x='x', y='y', data=data)
plt.title('Gráfico de Dispersión con Estilo ggplot')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
````

## Gráfico de distribución con seaborn y estilo ggplot

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
data = np.random.normal(size=100)

# Configurar seaborn para usar el estilo ggplot
sns.set(style="whitegrid")

# Crear el gráfico de distribución
sns.histplot(data, kde=True)
plt.title('Gráfico de Distribución con Estilo ggplot')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
````

# Visualización de Gráficos Interactivos en Visual Studio Code con Plotly

Cuando trabajamos en Visual Studio Code, puede ser complicado visualizar gráficos interactivos directamente en el editor. Sin embargo, podemos configurar `plotly` para que utilice el renderizador `browser`, lo que abrirá los gráficos en tu navegador web predeterminado. Aquí te explicamos cómo hacerlo.

## Configuración del Renderizador de Plotly para el Navegador

### Paso 1: Configurar el Renderizador

Primero, debemos configurar `plotly` para que utilice el renderizador `browser`. Esto se hace fácilmente al inicio de tu script de Python:

```python
import plotly.io as pio
pio.renderers.default = 'browser'


## Configurar el Renderizador del Navegador

````python
import plotly.io as pio
pio.renderers.default = 'browser'

import plotly.express as px
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el gráfico interactivo
fig = px.line(x=x, y=y, title='Gráfico Interactivo de Líneas')
fig.update_xaxes(title_text='Eje X')
fig.update_yaxes(title_text='Eje Y')
fig.show()
````

## Crear un Mapa Interactivo con Plotly y GeoPandas

Los mapas interactivos son una excelente manera de visualizar datos geoespaciales. A continuación, te mostramos cómo crear un mapa interactivo usando `plotly` y `geopandas` en Python.

## Pasos para Crear un Mapa Interactivo

### Paso 1: Instalar las Dependencias

Asegúrate de tener instaladas las bibliotecas necesarias. Puedes instalar `plotly` y `geopandas` con pip:

```bash
pip install plotly geopandas


## Paso 2: Configurar Plotly para el Navegador

En tu script de Python, configura plotly para que use el renderizador browser. Esto abrirá el gráfico en tu navegador web predeterminado.

```Python
import plotly.io as pio
pio.renderers.default = 'browser'
```

## Paso 3: Cargar el Shapefile
Debes tener un archivo shapefile (SHP) con datos geoespaciales. Puedes descargar archivos shapefile de fuentes como Natural Earth. En este ejemplo, usamos un shapefile que contiene información sobre países:

```Python
import geopandas as gpd

# Cargar el shapefile desde la carpeta local
world = gpd.read_file("ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp")

```

## Paso 4: Crear el Mapa Interactivo
Usa plotly.express para crear el mapa interactivo. En este ejemplo, vamos a crear un mapa de coropletas, que visualiza los países coloreados por continente.

```Python

import plotly.express as px

# Crear el gráfico interactivo
fig = px.choropleth(world, 
                    geojson=world.geometry, 
                    locations=world.index, 
                    color='CONTINENT',
                    title='Mapa Interactivo del Mundo por Continente')

# Ajustar la visualización del mapa
fig.update_geos(fitbounds="locations", visible=False)
```


# Paso 5: Mostrar el Mapa
Finalmente, usa el método show para abrir el mapa en tu navegador web.

```Python
fig.show()

```

````python
import plotly.io as pio
pio.renderers.default = 'browser'

import plotly.express as px
import geopandas as gpd

# Cargar el shapefile desde la carpeta local
world = gpd.read_file("ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp")

# Crear el gráfico interactivo
fig = px.choropleth(world, 
                    geojson=world.geometry, 
                    locations=world.index, 
                    color='CONTINENT',
                    title='Mapa Interactivo del Mundo por Continente')

# Ajustar la visualización del mapa
fig.update_geos(fitbounds="locations", visible=False)

# Mostrar el mapa
fig.show()
````

# 3D Surface Plots in Python


````python
import plotly.graph_objects as go
import pandas as pd
import requests
from io import StringIO

# Download data using requests
url = 'https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv'
response = requests.get(url)
z_data = pd.read_csv(StringIO(response.text))

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
````

# Ejemplos de Ploty

Passing x and y data to 3D Surface Plot


````python
url = 'https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv'
response = requests.get(url)
z_data = pd.read_csv(StringIO(response.text))

z = z_data.values
sh_0, sh_1 = z.shape
x, y = np.linspace(0, 1, sh_0), np.linspace(0, 1, sh_1)
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
````

````python
# Read data from a csv
z_data = pd.read_csv(StringIO(response.text))

fig = go.Figure(data=[go.Surface(z=z_data.values)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  scene_camera_eye=dict(x=1.87, y=0.88, z=-0.64),
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90)
)

fig.show()
````
