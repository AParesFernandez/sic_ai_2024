Código en español explicado

```python
import scipy.stats as st
import numpy as np

# 1.1. Distribución Normal Estándar
mu = 0  # Media de la distribución normal estándar
sigma = 1  # Desviación estándar de la distribución normal estándar

# Cuantil para un nivel de confianza del 95%
alpha = 0.95
z_alpha = st.norm.ppf(alpha, loc=mu, scale=sigma)
print("Cuantil para un nivel de confianza del 95%: ", z_alpha)

# 1.2. Distribución t de Student
alpha = 0.95
for df in [10, 50, 100, 2000, 100000]:
    t_alpha = st.t.ppf(alpha, df)
    print(f"Grados de libertad = {df}, Cuantil = {t_alpha:.6f}")

# 1.3. Distribución Chi-cuadrado
df = 5
x = 8
p_value = st.chi2.cdf(x, df)
print(f"Probabilidad de que una chi-cuadrado con 5 grados de libertad sea menor o igual a 8: {p_value:.6f}")

# Verificando la inversa de la CDF (función cuantil de distribución normal estándar)
alpha = p_value
chi2_alpha = st.chi2.ppf(alpha, df)
print(f"Cuantil para una probabilidad de {alpha:.6f} y 5 grados de libertad: {chi2_alpha:.6f}")
```

**Se ha confirmado que el cuantil es la función inversa de la función de distribución acumulada.**

Grafico de Densidad de Probabilidad

```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np

# Distribución Normal Estándar
mu = 0
sigma = 1
x = np.linspace(-3.5, 3.5, 100)  # Creamos un conjunto de valores para el eje x
y = st.norm.pdf(x, mu, sigma)  # Calculamos la densidad de probabilidad

# Distribución t de Student (ejemplo con 10 grados de libertad)
df = 10
t_x = np.linspace(-5, 5, 100)
t_y = st.t.pdf(t_x, df)

# Distribución Chi-Cuadrado (ejemplo con 5 grados de libertad)
chi2_x = np.linspace(0, 20, 100)  # Valores del eje x para chi-cuadrado deben ser positivos
chi2_y = st.chi2.pdf(chi2_x, df)

# Gráfica
plt.figure(figsize=(10, 6))

# Gráfica de la distribución normal
plt.plot(x, y, label="Normal Estándar", color="blue")

# Gráfica de la distribución t de Student
plt.plot(t_x, t_y, label="t de Student (df=10)", color="orange")

# Gráfica de la distribución Chi-Cuadrado
plt.plot(chi2_x, chi2_y, label="Chi-Cuadrado (df=5)", color="green")

# Etiquetas y título
plt.xlabel("Valor")
plt.ylabel("Densidad de Probabilidad")
plt.title("Visualización de Algunas Distribuciones de Probabilidad")
plt.legend()
plt.grid(True)

plt.show()
```

Explicación del grafico

**Visualización de distribuciones:** Nos permite visualizar la forma de cada distribución, su centro y su dispersión.

**Comparación de distribuciones:** Podemos comparar fácilmente las diferentes distribuciones y ver cómo se diferencian entre sí.

**Interpretación de probabilidades:** El área bajo la curva de una distribución de probabilidad representa la probabilidad de que una variable aleatoria tome un valor dentro de un cierto intervalo.
