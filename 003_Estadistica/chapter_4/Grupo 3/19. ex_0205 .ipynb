### 1. Interval Estimation of the Mean:

```python
import numpy as np
import scipy.stats as st
```

```python
# x contains a sample.
# n = sample size.
x = np.array([25,24,24,27,29,31,28,24,25,26,25,18,30,28,23,26,27,23,16,20,22,22,25,24, 24,25,25,27,26,30,25,25,26,26,25,24])
n = len(x)
```

```python
# SEM = Standard Error of the Mean.
# In order to do unbiased estimation of the standard deviation, we set ddof = 1.
x_mean = x.mean()
SEM = x.std(ddof=1)/np.sqrt(n*1.0)
```

#### 1.1 90% Confidence Interval:

```python
# Using the approximated quantiles of the Standard Normal.
(x_mean - 1.645*SEM, x_mean + 1.645*SEM)
```

```python
# Using the exact quantiles of the Standard Normal.
(x_mean - st.norm.ppf(0.95,loc=0,scale=1)*SEM, x_mean + st.norm.ppf(0.95,loc=0,scale=1)*SEM)
```

```python
# Using the interval() function from the SciPy library (Standard Normal).
st.norm.interval(0.9, loc = x_mean, scale = SEM )
```

```python
# Using the exact quantiles of the Student-t.
(x_mean - st.t.ppf(0.95,df=n-1)*SEM, x_mean + st.t.ppf(0.95,df=n-1)*SEM)
```

```python
# Using the interval() function from the SciPy library (Student-t).
st.t.interval(0.9, df = n-1, loc = x_mean, scale = SEM )
```

#### 1.2. 95% Confidence Interval:

```python
# Using the approximated quantiles of the Standard Normal.
(x_mean - 1.96*SEM, x_mean + 1.96*SEM)
```

```python
# Using the exact quantiles of the Standard Normal.
(x_mean - st.norm.ppf(0.975,loc=0,scale=1)*SEM, x_mean + st.norm.ppf(0.975,loc=0,scale=1)*SEM)
```

```python
# Using the interval() function from the SciPy library (Standard Normal).
st.norm.interval(0.95, loc = x_mean, scale = SEM )
```

```python
# Using the exact quantiles of the Student-t.
(x_mean - st.t.ppf(0.975,df=n-1)*SEM, x_mean + st.t.ppf(0.975,df=n-1)*SEM)
```

```python
# Using the interval() function from the SciPy library (Student-t).
st.t.interval(0.95, df = n-1, loc = x_mean, scale = SEM )
```

#### 1.3. 99% Confidence Interval:

```python
# Using the approximated quantiles of the Standard Normal.
(x_mean - 2.576*SEM, x_mean + 2.576*SEM)
```

```python
# Using the exact quantiles of the Standard Normal.
(x_mean - st.norm.ppf(0.995,loc=0,scale=1)*SEM, x_mean + st.norm.ppf(0.995,loc=0,scale=1)*SEM)
```

Código comentado en Español

```python
import numpy as np
import scipy.stats as st

# Datos de muestra
x = np.array([25,24,24,27,29,31,28,24,25,26,25,18,30,28,23,26,27,23,16,20,22,22,25,24, 24,25,25,27,26,30,25,25,26,26,25,24])
n = len(x)  # Tamaño de la muestra

# Cálculo de la media y el error estándar de la media
x_mean = x.mean()  # Calcula la media de los datos
SEM = x.std(ddof=1) / np.sqrt(n)  # Calcula el error estándar de la media

# Intervalo de confianza del 90%
# Usando cuantiles aproximados de la normal estándar
intervalo_90_aprox = (x_mean - 1.645*SEM, x_mean + 1.645*SEM)
print("Intervalo de confianza del 90% (aproximado):", intervalo_90_aprox)

# Usando cuantiles exactos de la normal estándar
intervalo_90_exacto = (x_mean - st.norm.ppf(0.95)*SEM, x_mean + st.norm.ppf(0.95)*SEM)
print("Intervalo de confianza del 90% (exacto):", intervalo_90_exacto)

# Usando la función interval() de SciPy (normal estándar)
intervalo_90_scipy_normal = st.norm.interval(0.9, loc=x_mean, scale=SEM)
print("Intervalo de confianza del 90% (SciPy, normal):", intervalo_90_scipy_normal)

# Usando cuantiles exactos de la t de Student
intervalo_90_t = (x_mean - st.t.ppf(0.95, df=n-1)*SEM, x_mean + st.t.ppf(0.95, df=n-1)*SEM)
print("Intervalo de confianza del 90% (t-Student):", intervalo_90_t)

# Usando la función interval() de SciPy (t-Student)
intervalo_90_scipy_t = st.t.interval(0.9, df=n-1, loc=x_mean, scale=SEM)
print("Intervalo de confianza del 90% (SciPy, t-Student):", intervalo_90_scipy_t)

# ... similarmente para intervalos de confianza del 95% y 99% ...
```

Histograma de la muestra con Distribución Normal Ajustada

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Datos de muestra
x = np.array([25,24,24,27,29,31,28,24,25,26,25,18,30,28,23,26,27,23,16,20,22,22,25,24, 24,25,25,27,26,30,25,25,26,26,25,24])

# Calculamos la media y el error estándar de la muestra
x_mean = x.mean()
SEM = x.std(ddof=1) / np.sqrt(len(x))

# Creamos los datos del histograma
counts, bins, patches = plt.hist(x, bins='auto', density=True, color='skyblue')

# Distribución normal teórica ajustada a los datos
mu = x_mean
sigma = SEM
plt.plot(bins, stats.norm.pdf(bins, mu, sigma), color='red', linewidth=2, label='Normal ($\mu=%f$, $\sigma=%f$)' % (mu, sigma))

# Etiquetamos el histograma
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.title('Histograma de la muestra con distribución normal ajustada')
plt.grid(True)
plt.legend()

# Mostramos el histograma
plt.show()
```

Explicación del grafico

**Histograma (barras azules):**

Las barras del histograma representan cómo se distribuyen los valores de tu muestra. Cada barra muestra la frecuencia (o densidad) de valores dentro de un rango específico.
En este caso, los valores son las respuestas de una muestra y las barras azules indican cuántas veces aparecen valores en esos rangos específicos.

**Curva de distribución normal (línea roja):**

La línea roja es una curva de distribución normal ajustada a los datos de tu muestra.
La distribución normal se ajusta utilizando la media **(
𝜇
μ)** y el error estándar **(
𝜎
σ)** de tu muestra.
La forma de esta curva muestra cómo se espera que se distribuyan los datos si siguen una distribución normal teórica basada en la media y el error estándar de tu muestra.
