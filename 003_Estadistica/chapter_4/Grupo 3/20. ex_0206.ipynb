### 1. Interval Estimation of the Proportion:

```python
import numpy as np
import scipy.stats as st
```

1). There are two candidates for the election: A and B.
The candidate A wants to survey his approval rating.
Out of 100 suerveyed, 55 answered positively.
Can A be sure of this election?
- Assume that A gets elected with 50% or more.
- Use 95% confidence interval.

```python
# 95% Confidence Interval.
# We apply the expression for the standard error (SE) from the lecture.
# W ignore additional corrections.
n_yes = 55
n_total = 100
p_mean = n_yes/n_total
SE = np.sqrt(p_mean*(1-p_mean)/n_total)
st.norm.interval(0.95, loc = p_mean, scale=SE)
```

*=> The confidence interval has the lower blound below 50%. So A cannot be sure with the confidence level of 95%.*

2). Out of 1000 suerveyed, 550 answered positively. Can A be sure of this election?
- Assume that A gets elected with 50% or more.
- Use 95% confidence interval.

```python
# 95% Confidence Interval.
# We apply the expression for the standard error (SE) from the lecture.
# W ignore additional corrections.
n_yes = 550
n_total = 1000
p_mean = n_yes/n_total
SE = np.sqrt(p_mean*(1-p_mean)/n_total)
st.norm.interval(0.95, loc = p_mean, scale=SE)
```

*=> The confidence interval has the lower bound above 50%. So A can be sure with the confidence level of 95%.*

Comentarios del código en español

```python
import numpy as np
import scipy.stats as st

# Ejercicio de estimación de intervalo para la proporción

# 1. Hay dos candidatos a la elección: A y B. El candidato A quiere encuestar su índice de aprobación. De 100 encuestados, 55 respondieron positivamente. ¿Puede A estar seguro de ganar la elección?
# Asumimos que A gana la elección con un 50% o más de votos.
# Usamos un intervalo de confianza del 95%.

n_si = 55  # Número de respuestas positivas
n_total = 100  # Tamaño total de la muestra
p_media = n_si / n_total  # Proporción de respuestas positivas
SE = np.sqrt(p_media * (1 - p_media) / n_total)  # Error estándar

# Intervalo de confianza del 95%
intervalo_confianza = st.norm.interval(0.95, loc=p_media, scale=SE)
print("Intervalo de confianza del 95%: ", intervalo_confianza)

# Si el límite inferior del intervalo es mayor que 0.5, A puede estar seguro de ganar con un 95% de confianza.

# 2. De 1000 encuestados, 550 respondieron positivamente. ¿Puede A estar seguro de ganar la elección?
# Asumimos que A gana la elección con un 50% o más de votos.
# Usamos un intervalo de confianza del 95%.

n_si = 550
n_total = 1000
p_media = n_si / n_total
SE = np.sqrt(p_media * (1 - p_media) / n_total)

# Intervalo de confianza del 95%
intervalo_confianza = st.norm.interval(0.95, loc=p_media, scale=SE)
print("Intervalo de confianza del 95%: ", intervalo_confianza)

# Si el límite inferior del intervalo es mayor que 0.5, A puede estar seguro de ganar con un 95% de confianza.
```

```python
import numpy as np
import scipy.stats as st
import seaborn as sns
import matplotlib.pyplot as plt

# Datos del primer caso
n_si_1 = 55
n_total_1 = 100
p_media_1 = n_si_1 / n_total_1
SE_1 = np.sqrt(p_media_1 * (1 - p_media_1) / n_total_1)
intervalo_confianza_1 = st.norm.interval(0.95, loc=p_media_1, scale=SE_1)

# Datos del segundo caso
n_si_2 = 550
n_total_2 = 1000
p_media_2 = n_si_2 / n_total_2
SE_2 = np.sqrt(p_media_2 * (1 - p_media_2) / n_total_2)
intervalo_confianza_2 = st.norm.interval(0.95, loc=p_media_2, scale=SE_2)

# Crear un DataFrame con los datos
data = {
    'Caso': ['Caso 1']*100 + ['Caso 2']*100,
    'Proporción': np.concatenate([np.random.normal(p_media_1, SE_1, 100), np.random.normal(p_media_2, SE_2, 100)])
}

# Crear el gráfico de violín
plt.figure(figsize=(10, 6))
sns.violinplot(x='Caso', y='Proporción', data=data)
plt.axhline(0.5, color='r', linestyle='--', label='50%')
plt.title('Distribución de la proporción de respuestas positivas')
plt.legend()
plt.show()
```

Esplicación del grafico

El gráfico de violín muestra la distribución de la proporción de respuestas positivas (índice de aprobación) para dos encuestas diferentes realizadas por el candidato A.

Caso 1: En la primera encuesta, A recibió respuestas positivas de 55 de 100 personas. La distribución de estas proporciones está representada en el gráfico por una forma que se asemeja a un violín. La parte más ancha del violín indica que esas proporciones son las más comunes en la distribución.

Caso 2: En la segunda encuesta, A recibió respuestas positivas de 550 de 1000 personas. Esta distribución también se representa por otra forma de violín en el gráfico.

La línea horizontal roja en el 50% representa el umbral necesario para que el candidato A gane la elección. Si la mayor parte del "violín" (la parte más ancha de la distribución) está por encima de esta línea, indica que es probable que A gane la elección.
