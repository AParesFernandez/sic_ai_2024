```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Datos de ejemplo
data = np.random.randn(100)

# Histograma con density=False
plt.hist(data, bins=30, density=False, alpha=0.2, label='Frecuencia')

# Histograma con density=True
plt.hist(data, bins=30, density=True, alpha=0.8, label='Densidad de probabilidad')

plt.legend()
plt.show()
```

```python
# Datos de ejemplo
data_country_a = np.random.normal(loc=170, scale=10, size=100)  # Alturas país A
data_country_b = np.random.normal(loc=170, scale=10, size=200)  # Alturas país B

# Histograma con density=True para país A y país B
plt.hist(data_country_a, bins=30, density=False, alpha=0.5, label='País A')
plt.hist(data_country_b, bins=30, density=False, alpha=0.5, label='País B')

plt.legend()
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.title('Distribución de Alturas')
plt.show()
```
