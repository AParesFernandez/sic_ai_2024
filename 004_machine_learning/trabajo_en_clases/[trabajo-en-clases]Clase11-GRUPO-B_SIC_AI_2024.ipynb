# Ejemplos de validación

##K-Fold Cross Validation
###Descripción:
K-Fold Cross Validation divide el conjunto de datos en
𝑘
k pliegues (o "folds"). El modelo se entrena
𝑘
k veces, cada vez utilizando
𝑘
−
1
k−1 pliegues para el entrenamiento y el pliegue restante para la validación. El proceso se repite
𝑘
k veces, con cada pliegue utilizado exactamente una vez para la validación.

Ventajas:

Mayor uso de los datos: Cada muestra se utiliza tanto para entrenamiento como para validación, lo que maximiza el uso de los datos disponibles.
Estimaciones más estables: Al promediar los resultados de
𝑘
k entrenamientos diferentes, se obtiene una estimación más estable y robusta del rendimiento del modelo.
Flexibilidad: Se puede ajustar el número de pliegues (
𝑘
k) para equilibrar el sesgo y la varianza.
Desventajas:

Costo computacional: Requiere entrenar el modelo
𝑘
k veces, lo que puede ser computacionalmente costoso, especialmente para modelos complejos o grandes conjuntos de datos.
Menor estimación de error: Puede subestimar el error de generalización cuando se usa con
𝑘
k pequeño.
##Leave-One-Out Cross Validation (LOO)
###Descripción:
Leave-One-Out Cross Validation es un caso especial de K-Fold donde el número de pliegues (
𝑘
k) es igual al número de muestras en el conjunto de datos. En cada iteración, se usa una sola muestra para la validación y el resto para el entrenamiento. Este proceso se repite tantas veces como muestras haya en el conjunto de datos.

Ventajas:

Máximo uso de datos para el entrenamiento: En cada iteración, casi todas las muestras se utilizan para el entrenamiento, lo que puede ser beneficioso para conjuntos de datos pequeños.
Estimación sin sesgo: Proporciona una estimación sin sesgo del error de generalización, ya que cada muestra se valida exactamente una vez.
Desventajas:

Costo computacional extremadamente alto: Requiere entrenar el modelo tantas veces como muestras haya en el conjunto de datos, lo que puede ser prohibitivo para grandes conjuntos de datos.
Alta varianza: Las estimaciones de error pueden tener alta varianza, ya que cada modelo se entrena con un conjunto de datos ligeramente diferente.
Ejemplo Comparativo
Para ilustrar las diferencias entre K-Fold y LOO, podemos realizar una comparación práctica usando un conjunto de datos. Aquí, utilizaremos el conjunto de datos Iris con un modelo de Random Forest:

```python
# from sklearn.<familia> import <Algoritmo>
```

```python
# Importar las library
from sklearn.model_selection import KFold, LeaveOneOut, cross_val_score
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
# Cargar el conjunto de datos Iris
iris = load_iris()
X, y = iris.data, iris.target
```

```python
iris
```

```python
# Crear el modelo instanciando y definiendo los hiperparámetros
model = RandomForestClassifier(random_state=42)
```

```python
# K-Fold Cross Validation instanciamos la técnica
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Realizar la validación cruzada
kf_scores = cross_val_score(model, X, y, cv=kf)
```

```python
print("Scores de K-Fold Cross Validation: ", kf_scores)
print("Media de los scores: ", np.mean(kf_scores))
```

```python
# Leave-One-Out Cross Validation instanciar
loo = LeaveOneOut()
```

```python
# usamos la técnica
loo_scores = cross_val_score(model, X, y, cv=loo)
```

```python
print(len(loo_scores))
```

```python
print("Scores de Leave-One-Out Cross Validation: ", loo_scores)
print("Media de los scores: ", np.mean(loo_scores))
```
