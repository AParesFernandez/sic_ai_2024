```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Cargar los datasets
dfIMDB = pd.read_csv('/content/IMDB-Movie-Data.csv')
dfMovies = pd.read_csv('/content/MoviesOnStreamingPlatforms.csv')
dfNetflix = pd.read_csv('/content/netflix_titles_nov_2019.csv')

# Renombrar columna en dfNetflix para hacer el merge
dfNetflix.rename(columns={'title': 'Title'}, inplace=True)

# Fusión de los datasets por la columna 'Title'
dfMerge = pd.merge(dfIMDB, dfNetflix, on='Title', how='inner')
dfFinal = pd.merge(dfMerge, dfMovies, on='Title', how='inner')

# Eliminar duplicados
dfFinal.drop_duplicates(inplace=True)

# Identificar columnas con valores nulos
null_values = dfFinal.isnull().sum() / len(dfFinal) * 100

# Imputación para columnas numéricas con la mediana
imputer_num = SimpleImputer(strategy='median')
dfFinal['Revenue (Millions)'] = imputer_num.fit_transform(dfFinal[['Revenue (Millions)']])
dfFinal['Metascore'] = imputer_num.fit_transform(dfFinal[['Metascore']])

# Codificación de la columna 'Age' con LabelEncoder en lugar de imputación
label_encoder = LabelEncoder()
dfFinal['Age'] = label_encoder.fit_transform(dfFinal['Age'])

# Codificación de la columna 'Type' con LabelEncoder (es binaria: Movie/TV Show)
dfFinal['Type'] = label_encoder.fit_transform(dfFinal['Type'])

# Transformación de la columna 'Genre' para One-Hot Encoding
dfFinal['Genre'] = dfFinal['Genre'].str.split(',')
dfFinal = dfFinal.explode('Genre')

# Aplicar One-Hot Encoding en la columna 'Genre'
dfFinal = pd.get_dummies(dfFinal, columns=['Genre'], prefix='Genre', drop_first=True)

# Resultado del preprocesamiento
dfFinal.head()
```

```python
import os
import zipfile
import requests
import pandas as pd

# URL del dataset
url = "https://files.grouplens.org/datasets/movielens/ml-20m.zip"
zip_file = "ml-20m.zip"
extract_dir = "ml-20m"

# Descargar el dataset
if not os.path.exists(zip_file):
    print("Descargando dataset...")
    response = requests.get(url)
    with open(zip_file, 'wb') as f:
        f.write(response.content)
    print("Descarga completa.")

# Descomprimir el dataset
if not os.path.exists(extract_dir):
    print("Descomprimiendo dataset...")
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print("Descompresión completa.")
```
